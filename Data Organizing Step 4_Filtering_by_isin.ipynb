```python
import pandas as pd
import numpy as np  
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_excel('customer_data.xlsx') # read the data
```

```python
male_data = data[data['Gender'] == 'Male'] # filter the data by male
male_data
```

```python
male_fre_data = male_data[male_data['Frequency_of_Purchases'] > 5]
male_fre_data
```

```python
churned_male_freq = male_fre_data.query('Churn_Status == "Yes"')    # filter the data by churn status
churned_male_freq
```

```python
churned_male_freq_city = churned_male_freq[churned_male_freq['City'].isin(['Chicago','Houston'])]
churned_male_freq_city
```
